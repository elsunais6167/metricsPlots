```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prr = pd.read_csv('packetRR.csv')

Vehicle = prr['Vehicles']
AODV = prr['AODV']
DSDV = prr['DSDV']
AODV_I = prr['AODV-Improved']
DSDV_I = prr['DSDV-Improved']

plt.figure(figsize=(10,5))

plt.plot(Vehicle, AODV, marker="x")
plt.plot(Vehicle, DSDV, marker="*")
plt.plot(Vehicle, AODV_I, marker="o")
plt.plot(Vehicle, DSDV_I, marker="d")

plt.title('Average Packet Recieve Rate')
plt.xlabel('Number of Vehicles')
plt.ylabel('Receive Rate')
plt.legend(['AODV no Scheduler', 'DSDV no Scheduler', 'AODV-Improved', 'DSDV-Improved'], loc='lower right')

plt.show()

```
